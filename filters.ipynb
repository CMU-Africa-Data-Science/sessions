```
img = cv2.imread('cat.jpg')
plt.imshow(img)
```

```
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
gray_img.shape
```

```
center = gray_img[60:78, 120:135]
print(center)

plt.imshow(center, cmap='gray')
```

```
median_filtered = cv2.medianBlur(gray_img, 3)
plt.imshow(median_filtered, cmap='gray')
```

```
for i, kernel_size in enumerate([3, 7, 11, 19]):
  filtered = cv2.medianBlur(gray_img, kernel_size)
  plt.subplot(2, 2, i+1)
  plt.imshow(filtered, cmap='gray')
```

```
edge_detection = cv2.Canny(gray_img, 0, 10)
plt.imshow(edge_detection, cmap='gray')
```
